**Reverse Complement function**

image.png

```python
DNA_Reverse_Complement = {
    'A' : 'T', # A becomes T
    'T' : 'A', # T becomes A
    'G' : 'C', # G becomes C
    'C' : 'G'  # C becomes G
}


def reverse_complement(seq):
  """We compute the complement first, then reverse the string"""
  return ''.join([DNA_Reverse_Complement[nucleotide] for nucleotide in seq])[::-1]  
```

```python
print('DNA String + Complement + Reverse Complement:\n')
print(f"5' {result3} 3'")
print(f"   {''.join(['|' for count in range(len(result3))])}")
print(f"3' {reverse_complement(result3)[::-1]} 5' [Complement]")
print(f"5' {reverse_complement(result3)} 3' [Reverse Complement]")
```

# Colorize the sequence for better representation

```python
def colored(seq):
    bcolors = {
        'A' : '\033[92m',
        'C' : '\033[94m',
        'G' : '\033[93m',
        'T' : '\033[91m',
        'U' : '\033[91m',
        'reset' : '\033[0;0m'
    }
    
    tmpStr = ""
    
    for nuc in seq:
        # if the key matches
        if nuc in bcolors:
            tmpStr += bcolors[nuc] + nuc
        else:
            tmpStr += bcolors['reset'] + nuc
            
    return tmpStr + '\033[0;0m'
```

```python
print('DNA String + Complement + Reverse Complement:\n')
print(f"5' {colored(result3)} 3'")
print(f"   {''.join(['|' for count in range(len(result3))])}")
print(f"3' {colored(reverse_complement(result3)[::-1])} 5' [Complement]")
print(f"5' {colored(reverse_complement(result3))} 3' [Reverse Complement]")
```

# GC Content Calculation

In molecular biology and genetics, GC-content is the percentage of nitrogenous bases in a DNA or RNA molecule that are either guanine or cytosine. This measure indicates the proportion of G and C bases out of an implied four total bases, also including adenine and thymine in DNA and adenine and uracil in RNA.

```python
def GC_Content_Cal(sequence):
    gc_count = sequence.count('G') + sequence.count('C')
    total_length = len(sequence)
    gc_content = gc_count / total_length * 100
    return gc_content
```

**Let's test the GC_Content_Cal function**

In this case we will find the GC Content from the RNA Sequence

```python
print(f'GC Content of DNA Sequence: {GC_Content_Cal(RNA_Seq)} %\n')
```

We need to implement a function that calculates GC Content in specific sub sets of a DNA sequence

```python
def GC_Content_subsec(seq, k=20):
    """GC Content in a DNA/RNA sub-sequence length k, 
    k = 20 by default. k is a representation of the sample size
    so change k based on your desired length.
    """
    res = []
    # start stop step
    for i in range(0, len(seq) - k + 1, k):
        subseq = seq[i:i + k] # select based on the value of k | start is i and end is k
        res.append(GC_Content_Cal(subseq)) # based on the selected length, perform the GC content caluclation
    return res
    
```

# Translation, Codon Usage

AThe RNA codon table is used to translate the sequence of nucleotides in an RNA molecule into the sequence of amino acids that make up a protein.

During translation, the RNA molecule is read in groups of three nucleotides called codons. Each codon corresponds to a specific amino acid, and the sequence of codons in the RNA molecule determines the sequence of amino acids in the resulting protein. The RNA codon table is a reference that lists the corresponding amino acid for each possible codon.

To use the RNA codon table to translate an RNA sequence, you would start at the first codon in the sequence and look up the corresponding amino acid in the table. Then you would move on to the next codon and repeat the process until you have translated the entire sequence. The resulting sequence of amino acids is the primary structure of the protein encoded by the RNA molecule.

The DNA codon table is used to translate the sequence of nucleotides in a DNA molecule into the sequence of amino acids that make up a protein. However, the DNA molecule itself is not directly used to synthesize proteins. Instead, the genetic information contained in DNA is first transcribed into RNA, which is then translated into protein.

During transcription, an enzyme called RNA polymerase reads the DNA sequence and synthesizes an RNA molecule that is complementary to one of the strands of the DNA. The RNA molecule produced during transcription is usually single-stranded and can be used to translate the genetic information encoded in the DNA into proteins.

So to use the DNA codon table to translate a DNA sequence into a protein, you would first need to transcribe the DNA sequence into RNA using RNA polymerase. Then you could use the RNA codon table to translate the RNA sequence into a protein.

We search for codon triplets(3) and swap them to their relative aminoacids using the DNA Codon dictionary.

```python
def translate_seq(seq, init_pos = 0):
    """Translate a DNA or RNA sequence into amino acids sequence"""
    
    # pos = sample of 3 nucleotide at once 
    
    #        Swap key and value pair                       start       stop      step
    return [RNA_Codons[seq[pos:pos + 3]] for pos in range(init_pos, len(seq) - 2, 3)]
```

To compute the frequency of each aminoacid present in the DNA sequence, we create a function that would count the occurrence of it. 

```python
from collections import Counter
```

```python
def codon_usage(seq, aminoacid, seq_type):
    """Provides the frequency of each codon 
    encoding a given aminoacid in the DNA or RNA sequence.
    Must Specify whether DNA or RNA with the seq_type parameter
    """
    
    tmpList = []
    for i in range(0, len(seq) - 2, 3): # 3 steps/iteration scan through sequence 
        # After swap in dictionary, if equals target aminoacid parameter then append the corresponding codon
        #triplet to tmpList = []
        if seq_type[seq[i:i + 3]] == aminoacid:
            tmpList.append(seq[i:i + 3])
            
    # keep track of count for triplets         
    freqDict = dict(Counter(tmpList))
    total = sum(freqDict.values()) # adds up all values in the key and value pairs 
    
    for seq in freqDict:
        freqDict[seq] = round(freqDict[seq] / total, 2) # divide the count of each triplets over the total amount of count for all triplets to obtain the frequency
        
    return freqDict
```

# Open Reading Frames

An open reading frame (ORF) is a continuous stretch of nucleotides in a DNA or RNA molecule that has the potential to be translated into a protein. ORFs are identified by starting at the beginning of the sequence and reading the nucleotides in groups of three, called codons, until a stop codon is encountered. An ORF is considered "open" if it starts with a start codon (usually AUG) and ends with a stop codon (usually UAA, UAG, or UGA). This is case the start codon is **'M'** and stop codon is **'_'** .

To obtain ORFs from an RNA sequence, you can use a software tool or script to search the sequence for potential ORFs. There are many tools available for this purpose, including ORF Finder, which is a web-based tool provided by the NCBI that allows you to search for ORFs in a DNA or RNA sequence.

To use ORF Finder, you would need to enter the RNA sequence as the query and specify the desired parameters, such as the minimum and maximum length of the ORFs to be identified. The tool will then search the sequence for ORFs and return a list of ORFs that meet the specified criteria. You can then use the information provided by ORF Finder to analyze the ORFs and determine their potential function and other characteristics.

The first reading frame begins at the first nucleotide of the sequence and reads the nucleotides in groups of three until the end of the sequence. The second reading frame begins at the second nucleotide of the sequence and reads the nucleotides in groups of three until the second-last nucleotide. The third reading frame begins at the third nucleotide of the sequence and reads the nucleotides in groups of three until the third-last nucleotide.

```python
def ORFs(seq):
    """
    Generate the 6 open reading frames of a DNA sequence and it's reverse complement 
    or the open reading frames of a mRNA sequence. This function will translate either
    sequence into amino acids sequence first, then append them into the open reading
    frames list
    """
    
    # Translate into amino acids then into reading frames
    frames = []
    frames.append(translate_seq(seq, 0)) # first nucleotide of the sequence
    frames.append(translate_seq(seq, 1)) # second nucleotide of the sequence
    frames.append(translate_seq(seq, 2)) # third nucleotide of the sequence
    
    return frames
```
